# 들여쓰기
    * 다른 언어와는 달리 제어문이나 함수 및 클래스에서 실행 코드 부분을 구분해주는 괄호 {}가 없음. 
    * 들여쓰기(indentation)으로 대신함.
    * space bar 또는 tab

```python
listdata = ['a', 'b', 'c']
if 'a' in listdata:
    print('a가 listdata에 있습니다.')
    print(listdata)
else:
    print('a가 listdata에 존재하지 않습니다.')
```

# if문 (if ~ else)

        if 조건:
            실행 코드1
        else:
            실행 코드2
    

```python
x = 1
y = 2

if x >= y:
    print('x가 y보다 크거나 같습니다.')
else:
    print('x가 y보다 작습니다.')
```

# if문 (if ~ elif)

        if 조건1:
            실행 코드1
        elif 조건2:
            실행 코드2
        else:
            실행 코드3

```python
x = 1
y = 2

if x > y:
    print('x가 y보다 크거나 같습니다.')
elif x < y:
    print('x가 y보다 작습니다.')
else:
    print('x와 y가 같습니다.')
```

# for문
        for 변수 in 범위:
            반복으로 실행할 코드
            
    * 문자열
    * 리스트나 튜플
    * 사전
    * 그 외 반복 가능한 객체

```python
scope = [1, 2, 3, 4, 5]
for x in scope:
    print(x)
```

```python
# 문자열
str = "python!!"
for c in str:
    print(c)
```

```python
# 리스트
list = [10,20,30,40,50]
for c in list:
    print(c)
```

```python
# 사전
ascii_codes = {'a':97, 'b':98, 'c':99}
for c in ascii_codes:
    print(c)
    print(ascii_codes[c])
```

```python
# range
for c in range(10):
    print(c)
```

# for문 (for ~ continue ~ break)

        for 변수 in 범위:
            ...
            if 조건1:
                continue # 조건1에 해당하면 다시 for문 처음으로
            ...
            else:
                break # for 반복문 탈출

```python
scope = [1, 2, 3, 4, 5]
for x in scope:
    print(x)
    if x < 3:
        continue
    else:
        break
```

# for문 (for ~ else)
        
        for 변수 in 범위:
            반복 실행 코드
        else:
            for 구문이 모두 실행되었을 때 실행할 코드
    
* else:로의 진입은 for문이 break 등에 의해 중간에 중단됨이 없이 정상적으로 실행이 다 되었을 경우

```python
scope = [1, 2, 3]
for x in scope:
    print(x)
    break
else:
    print('Perfect')
```
